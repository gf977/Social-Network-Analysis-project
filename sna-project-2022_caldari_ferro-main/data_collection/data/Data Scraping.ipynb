### Create a dictionary of the most followed
with user href as key and value: [name, number of booke read, number of friends, number of followers] 

```python
url= "https://www.goodreads.com/user_following/most_followed?country=IT&duration=a" #most followed italian readers
res = session.get(url, cookies = cookies, headers = dict(referer = url))
soup = BeautifulSoup(res.content, 'html.parser')
print(soup)
```

```python
soup = soup.select("td a")
```

```python
print(soup)
```

```python
d = {} #dictionary that will be used to build the csv
for user in range(0, len(soup)):
    l = str(soup[user]).split('"') #split <>
    if len(l) > 3:
        d[l[1]] = [l[3]] #creo chiave dizionario con il nome
    else:
        if l[1] in d.keys():
            info = l[2].split()[0][1::].replace(',', '') #recupero le info da mettere come valore della chiave
            #print(info)
            d[l[1]].append(info) #dictionary value
l = []
for i in d.keys():
    l.append(i)
    print('chiave:', i, ", valore: ", d[i])
```

### Retriving all the most followed friends

```python
for key in l: #for every most followed
    key = key[0:5]+key[10:len(key)] #remove show from the href
    #break
    url_user = 'https://www.goodreads.com/friend{}'.format(key) #href of one of the most followed friend
    print(url_user)
    res = session.get(url_user, cookies = cookies, headers = dict(referer = url))
    soup_f = BeautifulSoup(res.content, 'html.parser')
    #find the number of pages 
    page = soup_f
    x = page.find_all('a', attrs={'href': re.compile("skip_mutual_friends=false")})
    max_page = int(x[-2].get_text()) #numero pagine di amici
    print(max_page)
    for link in soup_f.find_all('a', attrs={'href': re.compile("/review/list/")}):# retrive all the user in that page an added in d
        try:
            num_book = int(link.get_text().split(' ')[0].replace(',', ''))
            #print(num_book)
        except:
            continue
        key_new = '/user/show'+link.get('href')[12::]
        #print(key)
        if key_new not in d:
            name = key_new.split('-')[1::]  
            name = ' '.join(map(str, name))
            #print(name)
            d[key_new] = [name, num_book, 0 , 0 ]
    for i in range(2, max_page+1):
        print(i)
        url_new = url_user + '?page={}&skip_mutual_friends=false'.format(i)
        #print(url_new)
        res = session.get(url_new, cookies = cookies, headers = dict(referer = url))
        soup_2 = BeautifulSoup(res.content, 'html.parser')
        for link in soup_2.find_all('a', attrs={'href': re.compile("/review/list/")}):# retrive all the user in that page an added in d
            #print(link)
            try:
                num_book = int(link.get_text().split(' ')[0].replace(',', ''))
                #print(num_book)
            except:
                continue
            key_new = '/user/show'+link.get('href')[12::]
            #print(key_new)
            if key_new not in d:
                name = key_new.split('-')[1::]  
                name = ' '.join(map(str, name))
            #print(name)
                d[key_new] = [name, num_book, 0 , 0 ]
   
```
